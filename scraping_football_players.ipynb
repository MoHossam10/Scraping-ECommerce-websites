```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
url="https://www.transfermarkt.com/fc-barcelona/kader/verein/131/saison_id/2023/plus/1"
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.9',
    'Accept-Encoding': 'gzip, deflate, br',
    'Referer': 'https://www.transfermarkt.com/',
    'Connection': 'keep-alive'
}
```

```python
response = requests.get(url,headers=headers)
response
```

```python
soup=BeautifulSoup(response.content, 'html.parser')
soup
```

```python
players_name=[]
odd_div=soup.find('div',class_='responsive-table').find('div',class_='grid-view').find('table',class_='items').findAll('tr',class_='odd')
even_div=soup.find('div',class_='responsive-table').find('div',class_='grid-view').find('table',class_='items').findAll('tr',class_='even')
for i in odd_div:
  playername=i.find('td',class_='posrela').find('table',class_='inline-table').find('tr').find('td',class_='hauptlink').find('a').text.strip()
  players_name.append(playername)

for j in even_div:
  playername=j.find('td',class_='posrela').find('table',class_='inline-table').find('tr').find('td',class_='hauptlink').find('a').text.strip()
  players_name.append(playername)



print(players_name)
print(len(players_name))
```

```python
players_nationality=[]
odd_div=soup.find('div',class_='responsive-table').find('div',class_='grid-view').find('table',class_='items').findAll('tr',class_='odd')
even_div=soup.find('div',class_='responsive-table').find('div',class_='grid-view').find('table',class_='items').findAll('tr',class_='even')
for i in odd_div:
  playernationality=i.find('img',class_='flaggenrahmen')
  # Check if the 'img' tag is found and extract the nationality from the 'alt' attribute
  if playernationality:
    nationality = playernationality['alt']
    players_nationality.append(nationality)

for j in even_div:
  playernationality=j.find('img',class_='flaggenrahmen')
  # Check if the 'img' tag is found and extract the nationality from the 'alt' attribute
  if playernationality:
    nationality = playernationality['alt']
    players_nationality.append(nationality)



print(players_nationality)
print(len(players_nationality))
```

```python
players_signedfrom=[]
odd_div=soup.find('div',class_='responsive-table').find('div',class_='grid-view').find('table',class_='items').findAll('tr',class_='odd')
even_div=soup.find('div',class_='responsive-table').find('div',class_='grid-view').find('table',class_='items').findAll('tr',class_='even')

for i in odd_div:
  zentriert=i.findAll('td',class_='zentriert')
  for k in zentriert:
    signedfrom=k.find('a')
    if signedfrom:
      team= signedfrom.img['alt']
      players_signedfrom.append(team)

for j in even_div:
  zentriert=j.findAll('td',class_='zentriert')
  for l in zentriert:
    signedfrom=l.find('a')
    if signedfrom:
      team= signedfrom.img['alt']
      players_signedfrom.append(team)

print(players_signedfrom)
print(len(players_signedfrom))
```

```python
players_image=[]
odd_div=soup.find('div',class_='responsive-table').find('div',class_='grid-view').find('table',class_='items').findAll('tr',class_='odd')
even_div=soup.find('div',class_='responsive-table').find('div',class_='grid-view').find('table',class_='items').findAll('tr',class_='even')

for i in odd_div:
  playerimage=i.find('td',class_='posrela').find('table',class_='inline-table').find('tr').find('img')
  if playerimage:
    image=playerimage['data-src']
    players_image.append(image)

for i in even_div:
  playerimage=i.find('td',class_='posrela').find('table',class_='inline-table').find('tr').find('img')
  if playerimage:
    image=playerimage['data-src']
    players_image.append(image)


print(players_image)
print(len(players_image))
```

```python
players_league=[]
players_team=[]
players_status=[]
for i in players_name:
  players_league.append('LaLiga')
  players_team.append('Barcelona')
  players_status.append('still playing')

print(players_league)
print(players_team)
print(players_status)
print(len(players_league))
print(len(players_team))
print(len(players_status))
```

```python
players_data = []

for i in range(len(players_name)):
    player_data = {
        'Name': players_name[i],
        'Status': players_status[i],
        'Nationality': players_nationality[i],
        'Signed From': players_signedfrom[i],
        'League': players_league[i],
        'team': players_team[i],
        'Image': players_image[i]
    }
    players_data.append(player_data)


csv_file = '/content/drive/MyDrive/Barcelona.csv'

with open(csv_file, 'w', newline='', encoding='utf-8-sig') as file:
    writer = csv.DictWriter(file, fieldnames=players_data[0].keys())
    writer.writeheader()
    writer.writerows(players_data)

print(f"Data saved to {csv_file} successfully.")
```
