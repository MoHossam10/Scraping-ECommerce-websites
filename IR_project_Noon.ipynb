```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
url = "https://www.noon.com/egypt-en/fashion/men-31225/shoes-17421/adidas/adidas_originals/?sort%5Bby%5D=popularity&sort%5Bdir%5D=desc"
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.9',
    'Accept-Encoding': 'gzip, deflate, br',
    'Referer': 'https://www.noon.com',
    'Connection': 'keep-alive'
}
```

```python
response = requests.get(url,headers=headers)
response
```

```python
soup=BeautifulSoup(response.content, 'html.parser')
soup
```

```python
product_div = soup.find('div', class_='sc-5d5b4ce7-7 jJqNra grid')
product_spans = product_div.find_all('span')

product_links = []
for span in product_spans:
  link = span.find('a')  
  if link:
    href = link['href']  
    product_page = "https://www.noon.com" + href
    product_links.append(product_page)



print(len(product_links))
print(product_links)
```

```python
product_div = soup.find('div', class_='sc-5d5b4ce7-7 jJqNra grid')
product_spans = product_div.find_all('span')

product_names = []
for span in product_spans:
  link = span.find('a')  
  if link:
    div1 = link.find('div',{'class':'sc-d6c0094a-0 rXgaS grid'}).find('div',{'class':'sc-d86a2d5b-0 hntgnU'}).find('div',{'class':'sc-d86a2d5b-18 kGIqQP'}).find('div',{'class':'sc-d86a2d5b-20 cYgkvD'})
    product_name = div1['title']
    product_names.append(product_name)
    


print(len(product_names))
print(product_names)
```

```python
product_div = soup.find('div', class_='sc-5d5b4ce7-7 jJqNra grid')
product_spans = product_div.find_all('span')

product_prices = []
for span in product_spans:
  link = span.find('a')  
  if link:
    div1 = link.find('div',{'class':'sc-d6c0094a-0 rXgaS grid'}).find('div',{'class':'sc-d86a2d5b-0 hntgnU'}).find('div',{'class':'sc-d86a2d5b-18 kGIqQP'}).find('div',{'class':'sc-d86a2d5b-21 krocCU'}).find('div',{'class':'sc-6073040e-0 ebUFcq'}).find('div',{'class':'sc-6073040e-1 jscnIS'}).findAll('strong',{'class':'amount'})
    for strong in div1:
      product_prices.append(strong.text)
print(len(product_prices))
print(product_prices)
```

```python
product_div = soup.find('div', class_='sc-5d5b4ce7-7 jJqNra grid')
product_spans = product_div.find_all('span')

product_ratings = []
for span in product_spans:
    link = span.find('a') 
    if link:
        try:
            div1 = link.find('div', {'class': 'sc-d6c0094a-0 rXgaS grid'}).find('div', {'class': 'sc-d86a2d5b-0 hntgnU'}).find('div', {'class': 'sc-d86a2d5b-18 kGIqQP'}).find('div', {'class': 'sc-d86a2d5b-22 fkiZXo'}).find('div', {'class': 'sc-d86a2d5b-23 hfOOnA'}).find('div', {'class': 'sc-d86a2d5b-24 leDpyA'}).find('div', {'class': 'sc-61515602-0 czLWQH'}).find('div', {'class': 'sc-e568c3b8-0 bYtWsd'})
            rating = div1.find('span', {'class': 'sc-e568c3b8-1 bFgxSY'}).text
            product_ratings.append(rating)
        except AttributeError:
            product_ratings.append("N/A")

print(len(product_ratings))
print(product_ratings)
```

```python
company_names = []

for product_name in product_names:
    
    first_word = product_name.split()[0]
    company_names.append(first_word)


print(len(company_names))
print(company_names)
```

```python
product_div = soup.find('div', class_='sc-5d5b4ce7-7 jJqNra grid')
product_spans = product_div.find_all('span')

product_images = []
for span in product_spans:
  link = span.find('a')  
  if link:
    div1 = link.find('div',{'class':'sc-d6c0094a-0 rXgaS grid'}).find('div',{'class':'sc-d86a2d5b-0 hntgnU'}).find('div',{'class':'sc-d86a2d5b-1 junKqF'}).find('div',{'data-qa'})

print(len(product_images))
print(product_images)
```

```python
product_div = soup.find('div', class_='sc-5d5b4ce7-7 jJqNra grid')
product_spans = product_div.find_all('span')

product_images = []
for span in product_spans:
    link = span.find('a')  
    if link:
        img = link.find('img')  
        if img and 'src' in img.attrs:
            image_url = img['src']  
            product_images.append(image_url)

print(len(product_images))
print(product_images)
```

```python
products_data = []

for i in range(len(product_names)):
    product_data = {
        'Name': product_names[i],
        'Price': product_prices[i],
        'Rating': product_ratings[i],
        'Company': company_names[i],
        'Image URL': product_images[i],
        'Product Link': product_links[i]
    }
    products_data.append(product_data)


csv_file = '/content/drive/MyDrive/IR_project_noon.csv'

with open(csv_file, 'w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=products_data[0].keys())
    writer.writeheader()
    writer.writerows(products_data)

print(f"Data saved to {csv_file} successfully.")
```
