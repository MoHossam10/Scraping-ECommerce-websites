```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
url="https://www.jumia.com.eg/nike/"
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.9',
    'Accept-Encoding': 'gzip, deflate, br',
    'Referer': 'https://www.jumia.com/',
    'Connection': 'keep-alive'
}
```

```python
response = requests.get(url,headers=headers)
response
soup=BeautifulSoup(response.content, 'html.parser')
soup
```

```python
product_div=soup.find('div',class_='-paxs row _no-g _4cl-3cm-shs')
product_articles=product_div.find_all('article')



product_link=[]
for article in product_articles:
  link=article.find('a')
  if link:
    href=link['href']
    product_page="https://www.jumia.com.eg/"+href
    product_link.append(product_page)
print(product_link)
print(len(product_link))
```

```python
product_div = soup.find('div', class_='-paxs row _no-g _4cl-3cm-shs')
product_articles=product_div.find_all('article')

product_names = []
for article in product_articles:
    link = article.find('a')
    if link:
        div1 = link.find('div', {'class': 'info'}).find('h3', {'class': 'name'})
        product_names.append(div1.text)

print(product_names)
print(len(product_names))
```

```python
product_div = soup.find('div', class_='-paxs row _no-g _4cl-3cm-shs')
product_articles=product_div.find_all('article')

product_price = []
for article in product_articles:
    link = article.find('a')
    if link:
        div1 = link.find('div', {'class': 'info'}).find('div', {'class': 'prc'})
        product_price.append(div1.text)
    
print(product_price)
print(len(product_price))
```

```python
product_div = soup.find('div', class_='-paxs row _no-g _4cl-3cm-shs')
product_articles=product_div.find_all('article')

product_rate = []
for article in product_articles:
    link = article.find('a')
    if link:
        div1 = link.find('div', {'class': 'info'}).find('div', {'class': 'rev'})
        if div1:
            product_rate.append(div1.text)
        else:
            product_rate.append('N/A')
    
print(product_rate)
print(len(product_rate))
```

```python
product_company = []

# Iterate over the product names
for product_name in product_names:
    # Extract the company name (first word of the product name)
    first_word = product_name.split()[0]
    product_company.append(first_word)


print(product_company)
print(len(product_company))
```

```python
product_div = soup.find('div', class_='-paxs row _no-g _4cl-3cm-shs')
product_articles=product_div.find_all('article')

product_images = []

for article in product_articles:
    link = article.find('a')
    if link:
        image_element = link.find('div', {'class': 'img-c'}).find('img')
        if image_element:
            image_url = image_element['data-src']
            product_images.append(image_url)
            
print(product_images)
print(len(product_images))
```

```python
products_data = []

for i in range(len(product_names)):
    product_data = {
        'Name': product_names[i],
        'Price': product_price[i],
        'Rating': product_rate[i],
        'Company': product_company[i],
        'Image URL': product_images[i],
        'Product Link': product_link[i]
    }
    products_data.append(product_data)


csv_file = '/content/drive/MyDrive/IR_project_jumia.csv'

with open(csv_file, 'w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=products_data[0].keys())
    writer.writeheader()
    writer.writerows(products_data)

print(f"Data saved to {csv_file} successfully.")
```
